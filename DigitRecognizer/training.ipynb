```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.utils import to_categorical
from keras.callbacks import EarlyStopping
from keras import layers
```

```python
# hyper param
epochs = 20
batch_num = 128
```

```python
# csv to DataFrame
train_path = 'DigitRecognizer/data/train.csv'
test_path = 'DigitRecognizer/data/test.csv'
submission_path = 'DigitRecognizer/data/sample_submission.csv'
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)
submission_df = pd.read_csv(submission_path)
```

```python
# separate trains, labels
x_train_df = train_df.iloc[:, 1:]
x_test_df = test_df.iloc[:, 0:]
y_train_df = train_df.iloc[:, 0]
```

```python
# df -> numpy
x_train = x_train_df.values
x_test = x_test_df.values
y_train = y_train_df.values
```

```python
# normalize 何度もやならい！
x_train = x_train / 255
x_test = x_test / 255
```

```python
# to categorical
y_train = to_categorical(y_train)
```

```python
x_test.shape
```

```python
x_train = x_train.reshape(42000, 28, 28, 1)
x_test = x_test.reshape(28000, 28, 28, 1)
```

```python
# model
model = Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))
model.add(layers.Dropout(0.25))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

model.summary()
```

```python
model.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['acc']
)
```

```python
history = model.fit(
    x_train, y_train,
    shuffle=True,
    epochs=epochs,
    batch_size=batch_num,
    validation_split=0.2
)
```

```python
loss = history.history['loss']
val_loss = history.history['val_loss']
acc = history.history['acc']
val_acc = history.history['val_acc']
xrange = range(len(loss))
plt.plot(xrange, loss, 'bo')
plt.plot(xrange, val_loss, 'b')
plt.legend()
plt.figure()
plt.plot(xrange, acc, 'bo')
plt.plot(xrange, val_acc, 'b')
plt.legend()
plt.show()
```

```python
prediction = model.predict(x_test)
```

```python
result = np.argmax(prediction, axis=1)
```

```python
imageIds = np.array(range(1, len(result)+1)).astype(int)
```

```python
imageIds
```

```python
result_df = pd.DataFrame(result, imageIds, columns=['Label'])
```

```python
result_df.to_csv('DigitRecognizer/data/img_recog2.csv', index_label=['ImageId'])
```
